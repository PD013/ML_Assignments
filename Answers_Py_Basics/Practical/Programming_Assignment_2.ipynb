# **1.Write a Python program to convert kilometers to miles?**
:

```python
km = int(input("Enter The distance in KM : "))
mile = 0.621371 * km

print(f"Distance in Km is{km} and in Miles is : {round(mile,4)} ")
# Here using round to limit the value after decimal till 4 only
```


# **2.Write a Python program to convert Celsius to Fahrenheit?**



```python
c = int(input("Enter the Temperature in Celsius :"))
f =(c * 9/5) + 32

print(f"Temperature in Celsius is :{c} and in Fahrenheit is :{f}")
```

# **4.Write a Python program to solve quadratic equation?**

```python
# (-b Â± (b ** 2 - 4 * a * c) ** 0.5) / (2 * a) Solution for quadratic equation
# Here **0.5 is under root
import cmath
# Module for Coomplex maths

a = 1
b = 5
c = 6

# calculate the discriminant
d = (b**2) - (4*a*c)

# find two solutions
sol1 = (-b-cmath.sqrt(d))/(2*a)
sol2 = (-b+cmath.sqrt(d))/(2*a)

print(f'The solution are {sol1} and {sol2}')
```

# **5.Write a Python program to swap two variables without temp variable?**

```python
a = 10
b = 20
print(f"This is the old a :{a} and b :{b}")
a, b = b, a
print(f"This is the new a :{a} and b :{b}")
```
