## **2.Write a Python Program to Display the multiplication Table?**


```python
def tables(base,cst):
    for x in range(1,cst):
        print(f'{base} X {x} = {base*x}')

num = int(input('Enter a number of which you want table: '))
cst = 11 # Here 11 is because of n-1 in range and 10 bcz tables are always till 10
tables(num,cst)
```

# **3.Write a Python Program to Print the Fibonacci sequence?**


```python
n = int(input('Enter the number of Fibonacci numbers you want: '))
fibonacci_sequence = [0, 1]

for i in range(2, n):
    next_number = fibonacci_sequence[i - 1] + fibonacci_sequence[i - 2]
    fibonacci_sequence.append(next_number)

print(f'The first {n} Fibonacci numbers are: {fibonacci_sequence}')

```

# **4.Write a Python Program to Check Armstrong Number?**


```python
num = int(input("Enter a number :- "))
num_of_digits = len(str(num))
temp = num
sum = 0
while temp>0:
  digit = temp % 10
  sum += digit **num_of_digits
  temp //= 10
if num == sum:
   print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
```

# **5.Write a Python Program to Find Armstrong Number in an Interval?**


```python
def is_armstrong_number(num):
    order = len(str(num))
    temp = num
    sum = 0
    while temp > 0:
        digit = temp % 10
        sum += digit ** order
        temp //= 10
    return sum == num

def find_armstrong_numbers_in_interval(start, end):
    armstrong_numbers = []
    for num in range(start, end + 1):
        if is_armstrong_number(num):
            armstrong_numbers.append(num)
    return armstrong_numbers


start_range = int(input("Enter the start of the range: "))
end_range = int(input("Enter the end of the range: "))


armstrong_numbers_list = find_armstrong_numbers_in_interval(start_range, end_range)


print("Armstrong numbers in the range {} to {}:".format(start_range, end_range))
print(armstrong_numbers_list)
```
