13. **What's the difference between tuples and lists?**



 Lists are `Mutable, Indexable and Slicable`. they can have values added, removed, or changed. Tuples are `Immutable` but `Indexable `and `Slicable`. the tuple values cannot be changed at all. Also, tuples are represented using parentheses,` ()`, while lists use the square brackets,` []`.

15. **How do you get a list value's tuple form? How do you get a tuple value's list form?**



```python
my_list = [1, 2, 3, 4]
print("This is the type of my_list",type(my_list))
tuple_form = tuple(my_list)
print("This is the Tuple Format coverted from list ",tuple_form)
print("This is the type of tuple_form",type(tuple_form))
list_form = list(tuple_form)
print("This is the List Format coverted from Tuple",list_form)
type(list_form)
```


17. **How do you distinguish between copy.copy() and copy.deepcopy()?**

The shallow copying creates shared references, while deep copying creates fully independent copies.

 In deep copy and copy.copy the changes does not reflect in the original list but in copy copy changes in `nested list` will reflect in original list and changes in deep copys `nested` or `non nested` list will not be reflected any where in original list
