
7. What is a shortcut for the following code?
   ```
   if 'color' not in spam:
       spam['color'] = 'black'
   ```


````python
spam.setdefault('color','black')
# setdefault() only adds a new key-value pair if the key does not already exist in the dictionary.
#  If the key is already present, the method will not modify the value associated with that key.
````
