
9. **What are the methods for right-justifying, left-justifying, and centering a string?**



```python
text = "Hello"
justified_text = text.rjust(10)  # Right-justifies the text within 10 spaces
print(justified_text)  # Output: '     Hello'

text = "Helloo"
justified_text = text.ljust(10,'_')  # Left-justifies the text within 10 spaces
print(justified_text)  # Output: 'Hello_____'

text = "Preet"
justified_text = text.center(11,'-')  # Left-justifies the text within 10 spaces
print(justified_text)  # Output: '---Preet---'
```

10. **What is the best way to remove whitespace characters from the start or end?**

 If you only want to remove whitespace from the start or end, you can use lstrip() (left strip) or rstrip() (right strip) and str.strip() to remove from Start and End simultaneously

```python
text2 = "  Hello , MY name is Preet  " #here 2 spaces in front and back
new1 = text2.strip()
print(new1)

text2 = "             Hello , MY name is Preet" #here are many spaces in front
new2 = text2.lstrip()
print(new2)

text3 = "Hello , MY name is Preet                     " #here 2 spaces in front and back
new3 = text3.rstrip()
print(new3)
```
